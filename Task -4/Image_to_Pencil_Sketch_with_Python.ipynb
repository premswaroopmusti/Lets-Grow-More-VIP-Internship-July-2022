```python
import cv2
import numpy as np
from PIL import Image
from IPython.display import display
image = cv2.imread('/content/drive/MyDrive/Lets_Grow_More_Internship/TASK_4/dsp.jpg')
```

```python
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
display(Image.fromarray(image))
```

```python
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

```python
inverted_image = cv2.bitwise_not(gray_image)
display(Image.fromarray(inverted_image))
```

```python
blurred_image = cv2.GaussianBlur(inverted_image, (21,21),0)
display(Image.fromarray(blurred_image))
```

```python
inverted_blurred = cv2.bitwise_not(blurred_image)
pencil_sktech = cv2.divide(gray_image, inverted_blurred, scale = 256.0)
display(Image.fromarray(pencil_sktech))
```
